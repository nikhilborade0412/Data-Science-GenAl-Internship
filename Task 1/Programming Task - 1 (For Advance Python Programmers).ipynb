### 1480. Running Sum of 1d Array

```python
class Solution(object):
    def runningSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        running = 0
        result = []
        for n in nums:
            running += n
            result.append(running)
        return result        
        
```

### 1470. Shuffle the Array

```python
class Solution(object):
    def shuffle(self, nums, n):
        """
        :type nums: List[int]
        :type n: int
        :rtype: List[int]
        """
        result = []
        for i in range(n):
            result.append(nums[i])     
            result.append(nums[i + n])  
        return result        
```
